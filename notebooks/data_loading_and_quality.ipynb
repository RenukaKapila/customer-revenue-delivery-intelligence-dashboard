```python
# Libraries
import pandas as pd
import sqlite3
import os
```

```python
# Create database folder if it does not exist
os.makedirs("../data/database", exist_ok=True)

# File paths
raw_path = "../data/raw/"
db_path = "../data/database/olist_ecommerce.db"
```

```python
customers = pd.read_csv(raw_path + "olist_customers_dataset.csv")
geolocation = pd.read_csv(raw_path + "olist_geolocation_dataset.csv")
order_items = pd.read_csv(raw_path + "olist_order_items_dataset.csv")
payments = pd.read_csv(raw_path + "olist_order_payments_dataset.csv")
reviews = pd.read_csv(raw_path + "olist_order_reviews_dataset.csv")
orders = pd.read_csv(raw_path + "olist_orders_dataset.csv")
products = pd.read_csv(raw_path + "olist_products_dataset.csv")
sellers = pd.read_csv(raw_path + "olist_sellers_dataset.csv")
category_translation = pd.read_csv(raw_path + "product_category_name_translation.csv")

print("All CSV files loaded successfully.")
```

```python
# Create SQLite database connection
conn = sqlite3.connect(db_path)

# Save each CSV as a SQL table
customers.to_sql("customers", conn, if_exists="replace", index=False)
geolocation.to_sql("geolocation", conn, if_exists="replace", index=False)
order_items.to_sql("order_items", conn, if_exists="replace", index=False)
payments.to_sql("payments", conn, if_exists="replace", index=False)
reviews.to_sql("reviews", conn, if_exists="replace", index=False)
orders.to_sql("orders", conn, if_exists="replace", index=False)
products.to_sql("products", conn, if_exists="replace", index=False)
sellers.to_sql("sellers", conn, if_exists="replace", index=False)
category_translation.to_sql("category_translation", conn, if_exists="replace", index=False)

conn.close()

print("SQLite database created successfully.")
```

```python
from pathlib import Path
import shutil

# Your main project path
BASE = Path(r"C:\Users\Renuka\Desktop\Customer Revenue & Delivery Intelligence Dashboard")

# 1. Create clean project folders
folders = [
    "data",
    "data/raw",
    "data/database",
    "notebooks",
    "sql",
    "dashboard",
    "reports",
    "images"
]

for folder in folders:
    (BASE / folder).mkdir(parents=True, exist_ok=True)

print("Folders checked/created.")


# 2. Move SQLite database files into data/database
db_locations = [
    BASE / "olist_ecommerce.db",
    BASE / "data" / "olist_ecommerce.db"
]

for db_path in db_locations:
    if db_path.exists():
        new_path = BASE / "data" / "database" / db_path.name
        
        if db_path != new_path:
            shutil.move(str(db_path), str(new_path))
            print(f"Moved database: {db_path.name} → data/database/")
        else:
            print(f"Database already in correct folder: {db_path.name}")


# 3. Make sure raw CSV files stay in data/raw
csv_files = list((BASE / "data").glob("*.csv"))

for csv_file in csv_files:
    new_path = BASE / "data" / "raw" / csv_file.name
    shutil.move(str(csv_file), str(new_path))
    print(f"Moved CSV: {csv_file.name} → data/raw/")


# 4. Delete all .ipynb_checkpoints folders
checkpoint_folders = list(BASE.rglob(".ipynb_checkpoints"))

for folder in checkpoint_folders:
    shutil.rmtree(folder)
    print(f"Deleted checkpoint folder: {folder}")

print("Checkpoint folders removed.")


# 5. Create important starter files if they do not exist
starter_files = {
    "README.md": "# Customer Revenue & Delivery Intelligence Dashboard\n\nProject overview will go here.\n",
    
    "requirements.txt": """pandas
numpy
matplotlib
scikit-learn
sqlite3
jupyter
""",
    
    "data_dictionary.md": "# Data Dictionary\n\nThis file explains the meaning of each dataset and important columns.\n",
    
    "reports/final_business_report.md": "# Final Business Report\n\n## Project Summary\n\n## Key Findings\n\n## Business Recommendations\n",
    
    "dashboard/dashboard_notes.md": "# Dashboard Notes\n\nThis file explains dashboard KPIs, visuals, and filters.\n"
}

for file_path, content in starter_files.items():
    full_path = BASE / file_path
    
    if not full_path.exists():
        full_path.write_text(content, encoding="utf-8")
        print(f"Created: {file_path}")
    else:
        print(f"Already exists: {file_path}")


print("\nProject structure is now cleaned and organized.")
```
